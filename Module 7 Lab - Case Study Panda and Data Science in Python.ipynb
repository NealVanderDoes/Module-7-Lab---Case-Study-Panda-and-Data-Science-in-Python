```python
import pandas as pd

# 1 Create
dataframe = pd.read_csv("telco_churn.csv")

# 1.2 Create from a Dictionary
tempdict = {"coll1":[1, 2, 3,], "coll2":[4, 5, 6], "coll3":[7, 8, 9]}
dictdf = pd.DataFrame.from_dict(tempdict)

# 2 Read
# 2.1 Show Top 5 and Bottom 5 Rows
print(dataframe.head(10))
print(dictdf.head())
print(dataframe.tail(15))

# 2.2 Show Column and Data Type
print(dataframe.columns)
print(dataframe.dtypes)

# 2.3 Summary Statistics
print(dataframe.describe())
print(dataframe.describe(include="object"))

# 2.4 Filtering Columns
print(dataframe.State)
print(dataframe["International plan"])
print(dataframe[["State", "International plan"]])
print(dataframe.Churn.unique())

# 2.5 Filtering Rows
print(dataframe[dataframe["International plan"] == "No"])
print(dataframe[(dataframe["International plan"] == "No") & (dataframe["Churn"] == True)])

# 2.6 Indexing with iloc
print(dataframe.iloc[14])
print(dataframe.iloc[14, -1])    
print(dataframe.iloc[22:33])

# 2.7 Indexing with loc
state = dataframe.copy()
state.set_index("State", inplace=True)
print(state.head())
print(state.loc["OH"])

# 3 Update
# 3.1 Dropping Rows
print(dataframe.isnull().sum())
dataframe.dropna(inplace=True)
print(dataframe.isnull().sum())

# 3.2 Dropping Columns
print(dataframe.drop("Area code", axis=1))

# 3.3 Creating Calculated Columns
dataframe["New Column"] = dataframe["Total night minutes"] + dataframe["Total intl minutes"]
print(dataframe.head())

# 3.4 Updating an Entire Column
dataframe["New Column"] = 100
print(dataframe.head())

# 3.5 Updating a Single Value
dataframe.iloc[0, -1] = 10
print(dataframe.head())

# 3.6 Condiation Based Updating using Apply
dataframe["Churn Binary"] = dataframe["Churn"].apply(lambda x: 1 if x==True else 0)
print(dataframe[dataframe["Churn"] == True].head())

# 4 Delete/Output
# 4.1 Output to CSV
dataframe.to_csv("output.csv")

# 4.2 Output to JSON
dataframe.to_json()

# 4.3 Output to HTML
dataframe.to_html()

# 4.4 Delete a Dataframe
del dataframe
```
